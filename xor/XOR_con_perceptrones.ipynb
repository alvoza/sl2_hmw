```python
import numpy as np
```

```python
def  operacion_or(x1,x2):
    parametros = [1,1,0]
    return  int(x1*parametros[0]  + x2*parametros[1] + parametros[2] > 0)
 
testx1 = 0
testx2 = 1
 
operacion_or(testx1,testx2)
```

```python
def  operacion_or_vect(x1,x2):
    parametros = [1,1,0]
    par_in = (x1, x2, 1)
    return  int(np.matmul(par_in, parametros) > 0)
 
testx1 = 1
testx2 = 0
 
operacion_or_vect(testx1,testx2)
```

```python
def  operacion_and(x1,x2):
    parametros = [1,1,-1]

    return  int(x1*parametros[0]  + x2*parametros[1] + parametros[2] > 0)
 
testx1 = 1
testx2 = 1
 
operacion_and(testx1,testx2)
```

```python
def  operacion_and_vect(x1,x2):
    parametros = [1,1,-1]
    par_in = (x1, x2, 1)
    return  int(np.matmul(par_in, parametros) > 0)
 
testx1 = 0
testx2 = 0
 
operacion_and_vect(testx1,testx2)
```

```python
def operacion_not(val):
    return int(bool(not val))
```

```python
def operacion_xor(x1,x2):
    temp1 = operacion_and_vect(x1,x2)
    temp2 = operacion_or_vect(x1,x2)
    nand_temp = operacion_not(temp1)
    temp_3 = operacion_and_vect(nand_temp,temp2)
 
    return temp_3
 
testx1 = 1
testx2 = 0
 
operacion_xor(testx1,testx2)
```
